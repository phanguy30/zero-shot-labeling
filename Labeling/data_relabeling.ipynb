```python
from pathlib import Path
DATA_ROOT = Path('dataset/cherry-detection-1')  
assert DATA_ROOT.exists(), f'Dataset folder not found: {DATA_ROOT}'
print('Dataset root:', DATA_ROOT.resolve())
```

```python
#Parsing the file to merge classes
for label_file in DATA_ROOT.rglob('*.txt'):
    with label_file.open('r') as f:
        lines = f.readlines()

    new_lines = []
    for line in lines:
        parts = line.strip().split()

        if not parts:
            continue

        parts[0] = '0'  # change class ID to 0

        

        new_lines.append(' '.join(parts)) # joining back the parts into a line

    with label_file.open('w') as f:
        f.write('\n'.join(new_lines) + '\n') #join lines and add a newline at the end
```
