```python
import cv2
import math
import numpy as np
from pathlib import Path


def make_image_grid(image_paths, grid_cols=4, img_size=None):
    images = []

    # -------------------------
    # Load images
    # -------------------------
    for p in image_paths:
        img = cv2.imread(str(p))
        if img is None:
            continue

        if img_size is not None:
            img = cv2.resize(img, img_size)

        images.append(img)

    if len(images) == 0:
        raise ValueError("No valid images")

    # -------------------------
    # Force same size
    # -------------------------
    h, w = images[0].shape[:2]
    images = [cv2.resize(img, (w, h)) for img in images]

    # -------------------------
    # Compute grid size
    # -------------------------
    grid_rows = math.ceil(len(images) / grid_cols)

    # -------------------------
    # Pad with blanks
    # -------------------------
    blank = np.zeros_like(images[0])
    while len(images) < grid_rows * grid_cols:
        images.append(blank)

    # -------------------------
    # Build grid (COLUMN-WISE)
    # -------------------------
    rows = []
    for r in range(grid_rows):
        row_imgs = []
        for c in range(grid_cols):
            idx = c * grid_rows + r   # 🔑 column-major indexing
            row_imgs.append(images[idx])
        rows.append(np.hstack(row_imgs))

    grid = np.vstack(rows)
    return grid


# =========================
# RUN
# =========================
if __name__ == "__main__":
    IMAGE_DIR = Path("/Users/peter/Documents/MDS_UBC/My Projects/zero-shot-labeling/compare/base_dino")
    OUTPUT_PATH = Path("base_dino_collage.jpg")

    image_paths = sorted(list(IMAGE_DIR.glob("*.jpg")))[:16]

    grid = make_image_grid(
        image_paths,
        grid_cols=4,
        img_size=(480, 480)  # or None if already same size
    )

    cv2.imwrite(str(OUTPUT_PATH), grid)

    print(f"Saved collage → {OUTPUT_PATH}")
```
