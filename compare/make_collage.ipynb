```python
import cv2
import math
import numpy as np
from pathlib import Path


def make_image_grid(image_paths, grid_cols=4, img_size=None):
    images = []

    for p in image_paths:
        img = cv2.imread(str(p))
        if img is None:
            continue

        # optional resize (keep None if you want original size)
        if img_size is not None:
            img = cv2.resize(img, img_size)

        images.append(img)

    if len(images) == 0:
        raise ValueError("No valid images")

    # ensure all images same size
    h, w = images[0].shape[:2]
    images = [cv2.resize(img, (w, h)) for img in images]

    grid_rows = math.ceil(len(images) / grid_cols)

    # pad with black images if needed
    blank = np.zeros_like(images[0])
    while len(images) < grid_rows * grid_cols:
        images.append(blank)

    # build grid
    rows = []
    for i in range(grid_rows):
        row = images[i * grid_cols:(i + 1) * grid_cols]
        rows.append(np.hstack(row))

    grid = np.vstack(rows)
    return grid


# =========================
# RUN
# =========================
IMAGE_DIR = Path("/Users/peter/Documents/MDS_UBC/My Projects/zero-shot-labeling/yolo_train_out/yolo_pt_manual_label_pred")
OUTPUT_PATH = Path("collage.jpg")

image_paths = sorted(list(IMAGE_DIR.glob("*.jpg")), reverse=True)[:16]

grid = make_image_grid(image_paths, grid_cols=4, img_size=(480, 480))

cv2.imwrite(str(OUTPUT_PATH), grid)

print(f"Saved collage → {OUTPUT_PATH}")
```
